## 1. Collecting Data from Word Files

```python
!pip install python-docx
```

```python
#Import library
from docx import Document
import docx
```

```python
#creating a word file object
doc = open("C:/Users/zaida/Downloads/Doc_NLP_Lab.docx","rb")
```

```python
document = docx.Document(doc)
```

```python
docu=""
for para in document.paragraphs:
    docu += para.text
```

```python
#to see the output call docu
print(docu)
```

## 2. Tokenization

```python
#Import library
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
#sentence tokenization
sentence = sent_tokenize(docu)
```

```python
#length of the sentences
print(len(sentence))
```

```python
#print the sentence
sentence
```

```python
#word tokenization
words = word_tokenize(docu)
```

```python
#length of the words
print(len(words))
```

```python
#print the words
print(words)
```

## 3. Frequency Distribution of the words

```python
#Import the libraries
from nltk.probability import FreqDist
```

```python
#Finding the frequency
freDis = FreqDist(words)
```

```python
#print 10 of the common words
freDis.most_common(20)
```

### 3.1 Plot the Frequency Graph

```python
#Import library
import matplotlib.pyplot as plt
```

```python
#Plot the graph for the words
freDis.plot(20)
```

### 3.2 Remove punctuation marks

```python
#empty list to store the words
wordPunc = []
```

```python
#remove the punctuation marks
for a in words:
    if a.isalpha():
        wordPunc.append(a.lower())
```

```python
#print the words without punctuation marks
print(wordPunc)
```

### 3.3 Plotting graph without Punctuation marks

```python
#find the frequency
wordNoPunc = FreqDist(wordPunc)
```

```python
#print 10 of the common words
wordNoPunc.most_common(20)
```

```python
#plot the graph for the words
wordNoPunc.plot(20)
```

### 3.4 List of stopwords

```python
#Import library
from nltk.corpus import stopwords

#list of the stop words
stopwords = stopwords.words("english")

#print the stopwords
print(stopwords)
```

### 3.5 Removing stopwords

```python
#empty list to store the clean words
CleanWords = []
```

```python
#remove the stopwords marks
for w in wordNoPunc:
    if w not in docu:
        CleanWords.append(w)

print(CleanWords)
```

```python
#length of the stopwords
print(len(CleanWords))
```

```python
#find the frequency
freDist = FreqDist(CleanWords)

#print 10 of the common words
freDist.most_common(10)
```

```python
#plot the graph for the words
freDist.plot(10)
```

### 5.3 Converting text to features using count vectorizing

```python
#Text
text = ["I am passionate in exploring python capabilities as it increase my knowledge and skill in Data Science"]
```

```python
#import the function
from sklearn.feature_extraction.text import CountVectorizer
```

## 6. WordNet

```python
#import library
from nltk.corpus import wordnet
```

```python
for words in wordnet.synsets("increase"):
    print(words)
```

```python
#how many differnet meanings
for words in wordnet.synsets("increase"):
    for lemma in words.lemmas():
        print(lemma)
```

```python
#all details for a word
word = wordnet.synsets("student")[0]
```

```python
#checking name
print(word.name())

#checking definition
print(word.definition())

#checking examples
print(word.examples)
```

### 6.1 Hyponym

```python
#using wordnet to find a hyponym
word = wordnet.synsets("watch")[0]

#find specific term
word.hyponyms()
```

```python
word = wordnet.synsets("watch")[0]

#get a name only
print(word.lemmas()[0].name())
```

### 6.2 Synonyms

```python
#empty list to store synonyms
synonyms = []

for words in wordnet.synsets("fun"):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())

synonyms
```

### 6.3 Antonyms

```python
#empty list to store antonyms
antonyms = []

for words in wordnet.synsets("easy"):
    for emma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())
antonyms
```

### 6.4 Finding the similarity between words

```python
#similarity in words
word1 = wordnet.synsets("ship","n")[0]
word2 = wordnet.synsets("boat","n")[0]

#cheking similarity
print(word1.wup_similarity(word2))
```
