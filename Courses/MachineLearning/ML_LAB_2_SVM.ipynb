##  1. Data Preprocessing

```python
# importing libraries  
import numpy as nm  
import matplotlib.pyplot as mtp  
import pandas as pd 
```

```python
#importing datasets  
df = pd.read_csv('Downloads/MSFT_forSVM.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

#### Get Independent and dependent data

```python
#Create empty lists to store the independent and dependent data
days = list()
adj_close_prices = list()
```

```python
#Get the dates and lists to store independent and dependent data
df_days = df.loc[:, 'Date']
df_adj_price = df.loc[:, 'Close']
```

```python
#Create the independent data set
for day in df_days:
    days.append( [int(day.split('/')[1])])
    
#Create the dependent data set
for adj_close_price in df_adj_price:
    adj_close_prices.append(float(adj_close_price))
```

```python
x = days
y = adj_close_prices
```

```python
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.3, random_state=0)  
#feature Scaling  
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test) 
```

## 2. Fitting the SVM classifier to the training set

```python
from sklearn.svm import SVR # "Support vector classifier"  
```

```python
classifier = SVR(kernel='linear', C=1000.0) 
```

```python
classifier.fit(x_train, y_train) 
```

## 3. Predicting the test set result

```python
y_pred=classifier.predict(x_test)
```

## 5. Visualizing the training set result

```python
from matplotlib.colors import ListedColormap  
```

```python
x_set, y_set = x_train, y_train  
x1, x2 = nm.meshgrid(nm.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step  =0.01),  
                     nm.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01)) 
mtp.contourf(x1, x2, classifier.predict(nm.array([x1.ravel(), x2.ravel()] + [nm.repeat(0, x1.ravel().size) for _ in range(6)]).T).reshape(x1.shape),  
    alpha = 0.75, cmap = ListedColormap(('red', 'green')))  
mtp.xlim(x1.min(), x1.max())  
mtp.ylim(x2.min(), x2.max())  
for i, j in enumerate(nm.unique(y_set)):  
    mtp.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],  
        c = ListedColormap(('red', 'green'))(i), label = j)  
mtp.title('SVM classifier (Training set)')  
mtp.xlabel('Age')  
mtp.ylabel('Medical Predictor Variables')  
mtp.legend()  
mtp.show()
```
